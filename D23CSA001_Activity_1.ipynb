### Graphs for Activation Functions

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-15,15,100)
len(x)
```

```python
def sigmoid(X):
    return 1 / (1+np.exp(-X))
```

```python
y = sigmoid(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,label = "sigmoid curve", color = 'cyan')
plt.title("sigmoid curve")
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def ReLU(x):
    return np.maximum(0,x)
    
```

```python
y_relu = ReLU(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y_relu,label = "ReLU curve", color = 'red')
plt.title("ReLU curve")
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def leaky_relu(X, alpha=0.1):
    return np.where(X>= 0, X, alpha * X)
```

```python
y_leaky_relu = leaky_relu(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y_leaky_relu,label =" Leaky Relu", color = 'green')
plt.title("Leaky ReLU")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
y_tanh = np.tanh(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y_tanh,label = "Tanh", color = 'violet')
plt.title("Tanh")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
