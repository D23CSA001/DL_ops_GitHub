### Graphs for Activation Functions

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
given_random_values = [-3.5, -1.2, 0, 2.8, -4.1, 1.5, -0.7, 3.2, -2.4, 4.6]
```

```python
def sigmoid(X):
    return 1 / (1+np.exp(-X))
```

```python
y_sigmoid = []
```

```python
for value in given_random_values:
    y = sigmoid(value)
    y_sigmoid.append(y)
```

```python
y_sigmoid
```

```python
x = np.array(given_random_values)
x
```

Sorting all the values so it will be easy to plot the graph

```python
X_sort = np.sort(x)
```

```python
y_sigmoid = np.array(y_sigmoid)
y_sigmoid
```

```python
y_sigmoid_sort = np.sort(y_sigmoid)
```

### Sorting the values to plot the curves which is to be commited in the bug fix branch

```python
plt.figure(figsize=(8,6))
plt.plot(X_sort,y_sigmoid_sort,label = "Sigmoid curve", color = 'green')
plt.title("Sigmoid for current values")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

### Finding the given activation function values for given data point to commit in the feature 2 branch

ReLU

```python
def ReLU(x):
    return np.maximum(0,x)
```

```python
y_relu = []
```

```python
for value in given_random_values:
    y = ReLU(value)
    y_relu.append(y)
```

```python
print(y_relu)
```

```python
y_relu = np.array(y_relu)
```

```python
y_relu_sort = np.sort(y_relu)
```

```python
plt.figure(figsize=(8,6))
plt.plot(X_sort,y_relu_sort,label = 'ReLU on given values',color='cyan')
plt.title("ReLU on given random values")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

Leaky ReLU

```python
def leaky_relu(X, alpha=0.1):
    return np.where(X>= 0, X, alpha * X)
```

```python
y_leaky_relu = []
```

```python
for value in given_random_values:
    y = leaky_relu(value)
    y_leaky_relu.append(y)
```

```python
y_leaky_relu = np.array(y_leaky_relu)
y_leaky_relu
```

```python
y_leaky_relu_sort  = np.sort(y_leaky_relu)
```

```python
plt.figure(figsize=(8,6))
plt.plot(X_sort,y_leaky_relu_sort,label = 'Leaky ReLU on given values',color='green')
plt.title("Leaky ReLU on given random values")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
