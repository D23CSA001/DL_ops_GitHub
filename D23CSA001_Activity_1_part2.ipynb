### Graphs for Activation Functions

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
given_random_values = [-3.5, -1.2, 0, 2.8, -4.1, 1.5, -0.7, 3.2, -2.4, 4.6]
```

```python
def sigmoid(X):
    return 1 / (1+np.exp(-X))
```

```python
y_sigmoid = []
```

```python
for value in given_random_values:
    y = sigmoid(value)
    y_sigmoid.append(y)
```

```python
y_sigmoid
```

```python
x = np.array(given_random_values)
x
```

```python
y_sigmoid = np.array(y_sigmoid)
y_sigmoid
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y_sigmoid,label = "Sigmoid curve", color = 'green')
plt.title("Sigmoid for current values")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
